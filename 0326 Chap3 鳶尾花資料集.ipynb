```python
# -*- coding: utf-8 -*-
"""
Created on Thu Sep 24 16:17:13 2015

@author: Eddy_zheng
"""

from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# 具体函数方法可用 help(function) 查看，如：help(ax.plot_surface)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
from matplotlib import cm


def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-3, 3, 50)
ygrid = xgrid
x, y = np.meshgrid(xgrid, ygrid)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,
                y,
                f(x, y),
                rstride=2, cstride=2,
                cmap=cm.jet,
                alpha=0.7,
                linewidth=0.25)
ax.set_zlim(-0.5, 1.0)
plt.show()
```

```python

from sklearn import datasets,cluster,metrics
iris=datasets.load_iris()
iris_X=iris.data

result=cluster.KMeans(n_clusters=3).fit(iris_X) #分成 三類 
result.labels_

#算 silhouette score , 越接近1分群結果越好
silhouette_avg=metrics.silhouette_score(iris_X,result.labels_)
print(silhouette_avg)
```

```python
from sklearn import datasets,cluster,metrics
import matplotlib.pyplot as pt
iris=datasets.load_iris()
iris_X=iris.data

ks=range(2,15)
silhouette_avgs=[]

for n in ks :

    result=cluster.KMeans(n_clusters=n).fit(iris_X)
    silhouette_avg=metrics.silhouette_score(iris_X,result.labels_)
    silhouette_avgs.append(silhouette_avg)



pt.bar(ks,silhouette_avgs)
pt.show()
```
